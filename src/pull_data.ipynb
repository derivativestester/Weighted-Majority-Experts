```python
# reset f-date to datetime format and make sure all the values are in the first day of the month
def adjust_dates(group, threshold=3):
    # Sort dates
    sorted_dates = group.sort_values()
    # Initialize a new list to hold adjusted dates
    adjusted_dates = []
    
    # Use the first date as the starting point
    current_date = sorted_dates.iloc[0]
    
    for date in sorted_dates:
        # If the date is within the threshold, assign it to the current grouping date
        if (date - current_date).days <= threshold:
            adjusted_dates.append(current_date)
        else:
            # Update the current_date and add to adjusted_dates
            current_date = date
            adjusted_dates.append(current_date)
    
    # Return the adjusted dates corresponding to the original dates
    return pd.Series(adjusted_dates, index=group.index)

# Apply the function to each 'cik' group
mngrdata['adjusted_fdate'] = mngrdata.groupby('cik')['fdate'].transform(adjust_dates)

# get unique fdate count for each unique cik
# x-axis is cik, y-axis is fdate and datapoints are counts of fdate in our table
cik_fdate = mngrdata.pivot_table(index='cik', columns='adjusted_fdate', aggfunc='size', fill_value=0)
```
